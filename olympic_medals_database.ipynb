```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store 1st csv into dataframe
csv_file = "Resources/olympic_medal_dataset.csv"
olympic_med_data_df = pd.read_csv(csv_file)
olympic_med_data_df.head()
```

```python
# Store 2nd csv into dataframe
csv_file = "Resources/olympic_medal_dict.csv"
olympic_med_dict_df = pd.read_csv(csv_file)
olympic_med_dict_df.head()
```

```python
#Drop rows with NaN
olympic_med_data_df.dropna(inplace = True)
olympic_med_data_df.head()
```

```python
#Drop rows with NaN
olympic_med_dict_df.dropna(inplace = True)
olympic_med_dict_df.head()
```

```python
#Separate Country and country code in dict dataframe
olympic_med_dict2_df = olympic_med_dict_df["Team(IOC Code)"].str.strip().str.split(pat="\s", expand=True)[[0, 1]]
olympic_med_dict2_df.head()
```

```python
#Add new column to dataframe
country = olympic_med_dict2_df[0]
olympic_med_dict_df["Country"] = country
olympic_med_dict_df.head()
```

```python
#Reorder df
reorder_med_dict_df = olympic_med_dict_df[["Country", "Team(IOC Code)", "summer_rank", "s_gold", "s_silver", "s_bronze", "summer_total", "winter_rank", "w_gold", "w_silver", "w_bronze", "winter_total", "overall_rank", "total_gold", "total_silver", "total_bronze", "total_medals"]]
reorder_med_dict_df.head()
```

```python
#Delete Team(IOC Code)
del reorder_med_dict_df['Team(IOC Code)']
reorder_med_dict_df.head()
```

```python
#Connect to local database
connection_string = "postgres:postgres@localhost:5432/olympic_medals"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
reorder_med_dict_df.to_sql(name='olympic_medal_dict', con=engine, if_exists='append', index=False)
```

```python
olympic_med_data_df.to_sql(name='olympic_medal_data', con=engine, if_exists='append', index=False)
```
